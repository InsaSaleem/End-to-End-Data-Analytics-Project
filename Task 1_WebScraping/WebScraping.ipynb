##Step 1 вЂ” Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

##Step 2 вЂ” Scrape the Website

```python
# Storage lists
titles = []
prices = []
ratings = []
availability = []

# Rating words to numbers
rating_map = {
    'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5
}

base_url = "http://books.toscrape.com/catalogue/page-{}.html"

# Scrape first 5 pages (you can increase this)
for page in range(1, 6):
    print(f"Scraping page {page}...")
    response = requests.get(base_url.format(page))
    soup = BeautifulSoup(response.text, 'html.parser')

    books = soup.find_all('article', class_='product_pod')

    for book in books:
        # Title
        title = book.h3.a['title']
        titles.append(title)

        # Price
        price = book.find('p', class_='price_color').text.strip()
        prices.append(float(price.replace('Г‚', '').replace('ВЈ', '')))

        # Rating
        rating_word = book.p['class'][1]
        ratings.append(rating_map.get(rating_word, 0))

        # Availability
        avail = book.find('p', class_='instock').text.strip()
        availability.append(avail)

    time.sleep(1)  # Be polite вЂ” don't overload the server

print(f"\nвњ… Done! Scraped {len(titles)} books.")
```

##Step 3 вЂ” Save to DataFrame & CSV

```python
df_books = pd.DataFrame({
    'Title': titles,
    'Price (ВЈ)': prices,
    'Rating (1-5)': ratings,
    'Availability': availability
})

print(df_books.head(10))
print(f"\nShape: {df_books.shape}")

# Save to CSV
df_books.to_csv('books_scraped.csv', index=False)
print("вњ… Saved as books_scraped.csv")
```

##Step 4 вЂ” Quick Summary

```python
print("=== Dataset Summary ===")
print(df_books.describe())
print("\nTop 5 Cheapest Books:")
print(df_books.nsmallest(5, 'Price (ВЈ)')[['Title', 'Price (ВЈ)', 'Rating (1-5)']])
print("\nTop 5 Highest Rated Books:")
print(df_books.nlargest(5, 'Rating (1-5)')[['Title', 'Price (ВЈ)', 'Rating (1-5)']])
```
